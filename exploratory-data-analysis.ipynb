### **II. Handling Outliers**
---
Does any of the numerical features has outliers? Hint(Histograms, Distributions, Box Plots)

```python
plt.figure(figsize=(20,15))
sns.boxplot(data=data)
```

```python
sns.boxplot(y='engine-size',data=data)
```

```python
q3=data['engine-size'].describe()[6]
q1=data['engine-size'].describe()[4]
outliers=q3+1.5*(q3-q1)
```

```python
# delete value oultliers from column engine-size
len(data[data['engine-size'] > outliers])
# there are 10 outliers value
```

```python
data.drop(data[data['engine-size'] > outliers].index,inplace=True)
```

```python
sns.boxplot(y='engine-size',data=data)
```

```python
# success drop raws outliers from engine-size column
```

```python
sns.boxplot(y='wheel-base',data=data)
```

```python
q3=data['wheel-base'].describe()[6]
q1=data['wheel-base'].describe()[4]
outliers=q3+1.5*(q3-q1)
data.drop(data[data['wheel-base'] > outliers].index,inplace=True)
```

```python
sns.boxplot(y='wheel-base',data=data)
```

```python
# question comparison between cars about price
df=data.groupby('make')['price'].agg('mean')
df=pd.DataFrame(df)
df.reset_index(inplace=True)
data_types_car=df.sort_values(by='price',ascending=False)
plt.figure(figsize=(26,6))
plt.bar('make','price',data=data_types_car)
plt.title("comparison between type of cars about price")
plt.xlabel("types of Cars")
plt.ylabel("Average Price")
```

```python
# how does type of fuel affect price?
df=data.groupby('fuel_type')['price'].agg('mean')
df=pd.DataFrame(df)
df.reset_index(inplace=True)
plt.bar(df['fuel_type'],df['price'])
plt.title("comparison between number of doors per price")
plt.xlabel("types of fuel")
plt.ylabel("price of each type")
```
